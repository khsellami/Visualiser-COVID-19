```python
#Charger les données COVID-19
import pandas as pd

# Charger les données depuis un fichier CSV ou une source en ligne
data = pd.read_csv("covid_data.csv")

# Afficher les premières lignes pour comprendre la structure des données
print(data.head())

```

```python
#Préparer les données
data['Date'] = pd.to_datetime(data['Date'])  # Convertir en format datetime
# worldwide_data = data.groupby('Date')['Confirmed'].sum().reset_index()
# china_data = data[data['Country'] == 'China'].groupby('Date')['Confirmed'].sum().reset_index()
```

```python
#Visualiser les cas confirmés dans le monde
import matplotlib.pyplot as plt
import seaborn as sns

# Visualiser les cas confirmés dans le monde
plt.figure(figsize=(10, 6))
# sns.lineplot(x='Date', y='Confirmed', data=worldwide_data, label='Worldwide', color='blue')
# sns.lineplot(x='Date', y='Confirmed', data=china_data, label='China', color='red')

plt.title('Confirmed COVID-19 Cases Worldwide vs China')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Comparer la Chine avec d'autres pays
countries = ['China', 'Italy', 'USA', 'India']
# country_data = data[data['Country'].isin(countries)]

plt.figure(figsize=(12, 7))
# sns.lineplot(x='Date', y='Confirmed', hue='Country', data=country_data)

plt.title('COVID-19 Confirmed Cases in Selected Countries')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend(title='Country')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Ajouter une échelle logarithmique
plt.figure(figsize=(10, 6))
# sns.lineplot(x='Date', y='Confirmed', data=worldwide_data, label='Worldwide', color='blue')
# sns.lineplot(x='Date', y='Confirmed', data=china_data, label='China', color='red')

plt.yscale('log')  # Ajouter une échelle logarithmique
plt.title('Confirmed COVID-19 Cases (Log Scale)')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases (log scale)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.show()
```
